* ###GRÁFICO FUNÇÃO GENÉRICA###

** USANDO O WINSTON**

```julia
using Winston
```

```julia
# vetor x definido como os pontos de 0 a 10 dividos de 0.5 em 0.5
# é necessário o ".*" para calcular elemento por elemento do vetor x

x = 0:0.5:10
y(x) = sin(x).*x

Winston.plot(x, y(x),"r")           
```

** Melhorando ainda mais**

```julia
# vetor x definido como os pontos de 0 a 10 dividos de 0.5 em 0.5
# é necessário o ".*" para calcular elemento por elemento do vetor x

x = 0:0.5:10
y(x) = sin(x).*x

# "-or" aplica a reta definida pela string "-" e a string "o" sobre os pontos
Winston.plot(x, y(x),"-or", color="red")

Winston.legend(["Sin(x)"])          # aplica a legenda
Winston.title("Gráfico do sin(x)")  # aplica a o título na parte superior (padrão)
Winston.xlabel("x")                 # aplica a nome ao eixo X
Winston.ylabel("y = sin(x)")        # aplica a nome ao eixo Y
Winston.xlim(0,4*pi)                # limita o tamanho do eixo x em 0 a 4*pi
Winston.ylim(-10,9)                 # limita o tamanho do eixo y em -10 a 9
Winston.grid()                      # aplica grade
```

** GERADOS AO MESMO TEMPO SOBRE A MESMA FIGURA**

** Usando o Winston**

```julia
using Winston
```

```julia
# vetor x definido como os pontos de 0 a 10 dividos de 0.5 em 0.5
# é necessário o ".*" para calcular elemento por elemento do vetor x

x = 0:0.5:10;
y = sin(x).*x;
g = cos(x).*x;
```

```julia
Winston.plot(x,y,"-or",x,g,"-*bk")

Winston.legend(["sin (x)"; "cos(x)"])       # aplica a legenda
Winston.title("Gráfico sin(x) e cos(x)")    # aplica a o título na parte superior (padrão)
Winston.xlabel("x")                         # aplica a nome ao eixo X
Winston.ylabel("y = sin(x)")                # aplica a nome ao eixo Y
Winston.xlim(0,4*pi)                        # limita o tamanho do eixo x em 0 a 4*pi
Winston.ylim(-10,9)                         # limita o tamanho do eixo y em -10 a 9
Winston.grid()                              # aplica grade
```

```julia
# vetor x definido como os pontos de 0 a 10 dividos de 0.5 em 0.5
# é necessário o ".*" para calcular elemento por elemento do vetor x

x = 0:0.5:10;
y = sin(x).*x;
g = cos(x).*x;

using Gadfly
set_default_plot_size(15cm, 12cm)
Gadfly.plot(

layer(x=x, y=y(x), Geom.point,Geom.line, Theme(default_color=color("green"))),
layer(x=x, y=g(x), Geom.point,Geom.line, Theme(default_color=color("red"))),

Scale.x_continuous(minvalue=0, maxvalue=4*pi),
Scale.y_continuous(minvalue=-8, maxvalue=10)
)
```

** Plotando superfícies **

```julia
x = linspace(-5, 5, 100)' # tem que ter a aspas simples no final

y = linspace(-5, 5, 100)

r = sqrt(x.^2 .+ y.^2)

z = cos(r)

surf = PyPlot.plot_surface(x, y, z)
```

 Melhorando o gráfico 3D 


```julia
x = linspace(-5, 5, 100)' # tem que ter a aspas simples no final

y = linspace(-5, 5, 100)

r = sqrt(x.^2 .+ y.^2)

z = cos(r)

surf = PyPlot.plot_surface(x, y, z, rstride=1, cstride=1, linewidth=0, antialiased=true, cmap="coolwarm")
```
