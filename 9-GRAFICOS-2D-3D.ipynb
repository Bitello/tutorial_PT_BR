** GERADOS AO MESMO TEMPO SOBRE A MESMA FIGURA**

** Usando o Winston**

```julia
# vetor x definido como os pontos de 0 a 10 dividos de 0.5 em 0.5
# é necessário o ".*" para calcular elemento por elemento do vetor x

x = 0:0.5:10;
y(x) = sin(x).*x;
g(x) = cos(x).*x;
```

```julia
using Winston
Winston.plot(x,y(x),"-or",x,g(x),"-*bk")

# '-or' -> coloca um círculo nos pontos (x,y) e liga os pontos com uma linha vermelha (red)
# 'LineWidth' espessura da linha 
# 'markersize' tamanho do circulo
# 'markerfacecolor' cor do círculo
# 'markeredgecolor' cor da borda do círculo

Winston.legend(["sin (x)"; "cos(x)"])       # aplica a legenda
Winston.title("Gráfico sin(x) e cos(x)")    # aplica a o título na parte superior (padrão)
Winston.xlabel("x")                         # aplica a nome ao eixo X
Winston.ylabel("y = sin(x)")                # aplica a nome ao eixo Y
Winston.xlim(0,4*pi)                        # limita o tamanho do eixo x em 0 a 4*pi
Winston.ylim(-10,9)                         # limita o tamanho do eixo y em -10 a 9
Winston.grid()                              # aplica grade
```

**Usando o Gadfly**

```julia
Gadfly.plot([y,g],0,4*pi)
```

```julia
using Gadfly
set_default_plot_size(15cm, 12cm)
Gadfly.plot(

layer(x=x, y=y(x), Geom.point,Geom.line, Theme(default_color=color("green"))),
layer(x=x, y=g(x), Geom.point,Geom.line, Theme(default_color=color("red"))),

Scale.x_continuous(minvalue=0, maxvalue=4*pi),
Scale.y_continuous(minvalue=-8, maxvalue=10)
)
```
