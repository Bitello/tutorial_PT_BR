** Funções Genéricas, Anônimas e Simbólicas**

### GRÁFICO FUNÇÃO GENÉRICA

sintaxe:
```julia
plot(x_vetor, f.(x_vetor), label = "nome", outras_opções) 
ou 
plot(f, x_vetor, label = "nome", outras_opções)
```

````julia
using Plots
plotlyjs()

# vetor x definido como os pontos de 0 a 10 dividos de 0.5 em 0.5
# é necessário o ".*" para calcular elemento por elemento do vetor x

vetor_x = 0:0.5:10
f1(x) = sin(x)*x

plot(vetor_x, f1.(vetor_x))           
````

** Melhorando ainda mais**

## GRÁFICOS MULTIPLOS 

### GRÁFICOS GERADOS AO MESMO TEMPO SOBRE A MESMA FIGURA 

```julia
plot(x1_vetor,y1_vetor,label = "nome", outras_opções )
plot!(x2_vetor,y2_vetor,label = "nome", outras_opções  )

plot(var_vetor,f(var_vetor),label = "nome", outras_opções )
plot!(var_vetor,g(var_vetor),label = "nome", outras_opções )
```

````julia
# é necessário o ".*" para calcular elemento por elemento do vetor x
vetor_x = 0 : 0.5 : 3*pi
fg1(x) = sin(x)*x
fg2(x) = cos(x)*x
````

````julia
using Plots
plotlyjs()

plot( vetor_x , fg1.(vetor_x), label = "sin(x)*x")
plot!(vetor_x , fg2.(vetor_x), label = "cos(x)*x", size = (400,300)) 
````

**Outra maneira**
```julia
plot([funcao1, funcao2], vetor_x, label = ["nome" "nome"])
```

````julia
using Plots
plotlyjs()

plot([fg1 , fg2], vetor_x, label = ["sin(x)*x" "cos(x)*x"], size = (400,300))
````

### GRÁFICOS SOBREPOSTOS GERADOS UM APÓS O OUTRO 

````julia
using Plots
plotlyjs()

vetor_x = 0:0.1:2*pi

fg3(x) = sin(x)

plot_mult = plot(vetor_x,fg3(vetor_x), label = "Sin(x)")

for i = 1:3    
    plot_mult = plot!(vetor_x,fg3(vetor_x + i), label= ("Sin(x+$i)"))
    
end
plot(plot_mult, size = (400,300))
````

### GRÁFICOS MÚLTIPLOS LADO A LADO

```julia
plot(p1, p2,..,pn layout = @layout([a b .. n]), opções)
```

````julia
# funções
fg4(x) = sin(x)*x
fg5(x) = cos(x)*x
````

** Horizontal**

````julia
using Plots
plotlyjs()

vetor_x = 0 : 0.5 : 3*pi

p1 = plot(vetor_x, fg4.(vetor_x))
p2 = plot(vetor_x, fg5.(vetor_x), color = (:red))

plot(p1, p2, layout = @layout([a b]), title = ["sin(x)*x" "cos(x)*x"], label = ["sin(x)*x" "cos(x)*x"], size =(600,300))
````

````julia
using Plots
plotlyjs()
# é necessário o ".*" para calcular elemento por elemento do vetor x
vetor_x = 0 : 0.5 : 3*pi

p1 = plot(vetor_x, fg4.(vetor_x), title = "sin(x)*x", label = "sin(x)*x")
p2 = plot(vetor_x, fg5.(vetor_x), color = (:red), title = "cos(x)*x" , label = "sin(x)*x")

plot(p1, p2, size = (600,300))
````

**Vertical**

````julia
using Plots
plotlyjs()

plot(p1, p2, layout = @layout([a ; b]), label = ["sin(x)*x" "cos(x)*x"], size = (700,500))
````

## GRÁFICO EQUAÇÃO IMPLÍCITA

````julia
using Plots,ImplicitEquations
plotlyjs()

eqv1(x,y) = 2*x + 3*y - 6
eqv2(x,y) = 3*x - 4*y - 12

plot(eqv1 == 0)
plot!(eqv2 == 0 , size = (400,300))
````

**Outra forma usando o SymPy**

````julia
using SymPy
@vars x y
````

````julia
eqv_1 = 2*x + 3*y - 6
eqv_2 = 3*x - 4*y - 12

plot_implicit(Or(Eq(eqv_1, 0), Eq(eqv_2, 0)), (x, -5, 5), (y, -5, 5));
````

## INTERATIVIDADE 

Julia permite através dos pacotes gráficos, construir gráficos interativos, no qual um controlador permite manipular dados da função e gerar outro gráfico. Os melhores pacotes para este recurso são o GR(melhor opção) e o PyPlot em função da velocidade na geração e consistencia do gráfico gerado.

````julia
using Plots, Interact
gr() # funciona melhor que PlotlyJS e PyPlot

@manipulate for Grau = 0:10, Coef  = -10:10
    plot(x -> Coef*x^Grau + Coef*x + 1,-10,10) # função anonima
end
````

````julia
using Plots, Interact
gr() # funciona melhor que PlotlyJS e PyPlot

@manipulate for Grau = 0:10
    
    vetor_x = linspace(-5, 5, 300)
    vetor_y = linspace(-5, 5, 300)

    zf(x,y) = x^Grau + y^Grau

    plot(vetor_x , vetor_y , zf.(vetor_x,vetor_y),label = "Linha", size = (400,300) )
end
````
