<strong>GRÁFICO DE UMA FUNÇÃO</strong>

```julia
workspace()
```

```julia
# vetor x definido como os pontos de 0 a 10 dividos de 0.5 em 0.5
# é necessário o ".*" para calcular elemento por elemento do vetor x

x = 0:0.5:10;
y = sin(x).*x;
```

```julia
using Winston

Winston.plot(x, y,"r")           
```

Melhorando ainda mais

```julia
using Winston

# "-or" aplica a reta definida pela string "-" e a string "o" sobre os pontos
Winston.plot(x, y,"-or", color="red")

legend(["Sin(x)"])          # aplica a legenda
title("Gráfico do sin(x)")  # aplica a o título na parte superior (padrão)
xlabel("x")                 # aplica a nome ao eixo X
ylabel("y = sin(x)")        # aplica a nome ao eixo Y
xlim(0,4*pi)                # limita o tamanho do eixo x em 0 a 4*pi
ylim(-10,9)                 # limita o tamanho do eixo y em -10 a 9
grid()                      # aplica grade
```

Usando o Gadfly

```julia
using Gadfly

set_default_plot_size(20cm, 12cm)

Gadfly.plot(
x = x,
y = y,
Geom.point,
Geom.line,
Theme(default_color=color("red")),
Scale.x_continuous(minvalue=0, maxvalue=12),
Scale.y_continuous(minvalue=-2, maxvalue=2)
)
```

```julia
using Gadfly

set_default_plot_size(20cm, 12cm)

y1(x) = sin(x)*x

Gadfly.plot(y1,0,120)
```

<strong>GRÁFICOS MULTIPLOS </strong>

<strong>Gerados ao mesmo tempo sobre a mesma figura</strong>

```julia
# vetor x definido como os pontos de 0 a 10 dividos de 0.5 em 0.5
# é necessário o ".*" para calcular elemento por elemento do vetor x

x = 0:0.5:10;
y = sin(x).*x;
g = cos(x).*x;
```

```julia
using Winston
Winston.plot(x,y,"-or",x,g,"-*bk")

# '-or' -> coloca um círculo nos pontos (x,y) e liga os pontos com uma linha vermelha (red)
# 'LineWidth' espessura da linha 
# 'markersize' tamanho do circulo
# 'markerfacecolor' cor do círculo
# 'markeredgecolor' cor da borda do círculo

legend(["sin (x)"; "cos(x)"])       # aplica a legenda
title("Gráfico sin(x) e cos(x)")    # aplica a o título na parte superior (padrão)
xlabel("x")                         # aplica a nome ao eixo X
ylabel("y = sin(x)")                # aplica a nome ao eixo Y
xlim(0,4*pi)                        # limita o tamanho do eixo x em 0 a 4*pi
ylim(-10,9)                         # limita o tamanho do eixo y em -10 a 9
grid()                              # aplica grade
```

Usando o Gadfly

```julia
using Gadfly
set_default_plot_size(20cm, 12cm)
Gadfly.plot(

layer(x=x, y=y, Geom.point,Geom.line, Theme(default_color=color("green"))),
layer(x=x, y=g, Geom.point,Geom.line, Theme(default_color=color("red"))),

Scale.x_continuous(minvalue=0, maxvalue=4*pi),
Scale.y_continuous(minvalue=-8, maxvalue=10)
)
```

```julia
y1(x) = sin(x)*x;
g1(x) = cos(x)*x;
```

```julia
Gadfly.plot([y1,g1],0,4*pi)
```

```julia
using PyPlot

X = linspace(-5, 5, 300);

Y = linspace(-5, 5, 300);

Z = cos(X.^2)+sin(Y.^2).*X;

PyPlot.plot3D(X,Y,Z)
```

```julia
using PyPlot

#fig = figure()
X = linspace(-5, 5, 100)'
Y = linspace(-5, 5, 100)
R = sqrt(X.^2 .+ Y.^2)
Z = cos(R)
surf = PyPlot.plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=0, antialiased=true, cmap="coolwarm")
zlim(-1.01,1.01)
ax = gca()
ax[:zaxis][:set_major_locator](matplotlib[:ticker][:LinearLocator](10))
ax[:zaxis][:set_major_formatter](matplotlib[:ticker][:FormatStrFormatter]("%.02f"))
#fig[:colorbar](surf, shrink=0.5, aspect=5)
```
